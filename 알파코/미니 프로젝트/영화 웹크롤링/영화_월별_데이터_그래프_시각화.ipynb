```python
kor = pd.read_excel('국내 2017_2021 5개년 월별 데이터.xlsx',parse_dates=['년월'])
foreign = pd.read_excel('해외 2017_2021 5개년 월별 데이터.xlsx',parse_dates=['년월'])
whole = pd.read_excel('전체 2017_2021 5개년 월별 데이터.xlsx',parse_dates=['년월'])
#년월 데이트타임으로 바꾸고 연도 컬럼 추가해준다음 연도별 

kor['월'] = kor['년월'].dt.month
foreign['월'] = foreign['년월'].dt.month
whole['월'] = whole['년월'].dt.month

whole_people =  whole.groupby('월').sum() #17년~21년 월별 관객수
whole_people = whole_people.reset_index()
foreign_people =  foreign.groupby('월').sum() #17년~21년 월별 관객수
foreign_people = foreign_people.reset_index()
kor_people =  kor.groupby('월').sum() #17년~21년 월별 관객수
kor_people = kor_people.reset_index()

fig = px.line(whole_people, x='월',y='관객수', title='17~21년 5개년 전체 월별 관객수', height = 500  , width = 1000)
fig.show()


ko = pd.read_excel('국내 2004_2021 18개년 월별 데이터.xlsx',parse_dates=['년월'])
fo = pd.read_excel('해외 2004_2021 18개년 월별 데이터.xlsx',parse_dates=['년월'])
wh = pd.read_excel('전체 2004_2021 18개년 월별 데이터.xlsx',parse_dates=['년월'])
#년월 데이트타임으로 바꾸고 연도 컬럼 추가해준다음 연도별 

ko['월'] = ko['년월'].dt.month
fo['월'] = fo['년월'].dt.month
wh['월'] = wh['년월'].dt.month


ko_people = ko.groupby('월').sum()[['관객수']]
ko_people = ko_people.reset_index()
fo_people = fo.groupby('월').sum()[['관객수']]
fo_people = fo_people.reset_index()
wh_people = wh.groupby('월').sum()[['관객수']]
wh_people = wh_people.reset_index()

# 국내, 해외 다 합친 전체 월별 관객수 
fig0 = px.line(wh_people, x='월',y='관객수', title='04~21년 18개년 전체 월별 관객수', height = 500  , width = 1000)
fig0.show()



kor_price = kor[['년월','매출액']]
foreign_price = foreign[['년월','매출액']]
kor_fo_price = pd.merge(kor_price, foreign_price, on='년월') #국내 해외 매출액 
kor_fo_price.columns = ['년월', '국내 매출액', '해외 매출액']


fig1 = px.line(kor,x='년월',y='매출액', title='국내 연월별 매출액 추세', height = 500  , width = 1000)
fig2 = px.line(foreign,x='년월',y='매출액', title='해외 연월별 매출액 추세', height = 500  , width = 1000)
# fig1.update_xaxes(type='category')
fig1.show()
fig2.show()

fig3 = px.line(kor_fo_price,x='년월',y=['국내 매출액','해외 매출액'], title='2017~2021년 국내 해외 연월별 매출액 비교', height = 500  , width = 1000)
fig3.show()

fig4 = px.bar(kor_fo_price,x='년월',y=['국내 매출액','해외 매출액'],barmode = 'group', title='2017~2021년 국내 해외 연월별 매출액 비교', height = 500  , width = 1000)
fig4.show()
```

```python
open_count = kor[['개봉편수', '년월']]

open_count = open_count.rename(columns={'개봉편수':'국내 개봉편수'})
fo = foreign[['개봉편수']].rename(columns={'개봉편수':'해외 개봉편수'})
wh = whole[['개봉편수']].rename(columns={'개봉편수':'전체 개봉편수'})

count = pd.concat([open_count, fo, wh],axis=1)

count1 = px.bar(count, x='년월', y=['국내 개봉편수', '해외 개봉편수'],title='17~21년 국내 해외 영화 개봉편수', barmode='group')
count2 = px.line(count, x='년월', y=['국내 개봉편수', '해외 개봉편수'], title='17~21년 국내 해외 영화 개봉편수')
count3 = px.line(count, x='년월', y=['전체 개봉편수'], title='17~21년 전체 개봉편수')
count1.show()
count2.show()
count3.show()
```

```python
fo_people = fo_people.rename(columns={'관객수':'해외 관객수'})
ko_people = ko_people.rename(columns={'관객수':'국내 관객수'})
ko_fo_people = pd.merge(fo_people, ko_people, how='inner', on=None)
fig9 = px.line(ko_fo_people, x='월',y=['해외 관객수', '국내 관객수'], title='5년동안의 해외국내 월별 관객수', height = 500  , width = 1000)
fig9.show()
fig10 = px.line(ko_fo_people, x='월',y='국내 관객수', title='2017~2021년 국내 월별 관객수', height = 500  , width = 1000)
fig10.show()
```
